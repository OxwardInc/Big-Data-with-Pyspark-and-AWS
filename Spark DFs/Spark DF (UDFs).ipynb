```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit, udf
from pyspark.sql.types import IntegerType
spark = SparkSession.builder.appName('Spark DF').getOrCreate()
```

```python
df = spark.read.options(header = 'True', inferSchema = 'True').csv('../data/OfficeData.csv')
df.show()
```

```python
def get_total_salary(salary, bonus):
    return salary + bonus

totalSalaryUDF = udf(lambda x, y: get_total_salary(x, y), IntegerType())

df.withColumn("total_salary", lit(100)).show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit, udf
from pyspark.sql.types import DoubleType
spark = SparkSession.builder.appName('Spark DF').getOrCreate()
```

```python
df = spark.read.options(header = 'True', inferSchema = 'True').csv('../data/OfficeData.csv')
df.show()
```

```python
def get_incr(state, salary, bonus):
  sum = 0
  if state == "NY":
    sum = salary * 0.10
    sum += bonus * 0.05
  elif state == "CA":
    sum = salary * 0.12
    sum += bonus * 0.03
  return sum

incrUDF = udf(lambda x,y,z: get_incr(x,y,z), DoubleType())

df.withColumn("increment", incrUDF(df.state, df.salary, df.bonus)).show()
```

```python
df = df.groupBy('course', 'gender', 'age').count()
```
